## Load Datachile File

```python
import pandas as pd
import numpy as np

#pd.options.display.max_rows = 300

df1 = pd.read_csv("dim_countries_new.csv")
df1
```

## Load My File

```python
df2 = pd.read_csv("my_dim_countries.csv")
df2
```

## Merge and Comparison

```python
def parenthesis(row,col):
    target = row[col].split("(")[0]
    if target[-1] == " ":
        target = target[:-1]
    return target

df3 = df2
df3["country_name"] = df3.apply(parenthesis, col = "country_name", axis = 1)

df = df3.merge(df1, left_on = "country_name", right_on = "es_country_name", how = "left")
df = df[["census_country_code", "country_name_x", "country_code"]]
df = df.fillna(256)
df["country_code"] = df["country_code"].astype(int)
df.to_csv("datachile_census_country_id.csv")
```
